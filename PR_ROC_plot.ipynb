```python
import pandas as pd
import matplotlib.pyplot as plt

distil_data = pd.read_csv('roc_curve_distil.csv')
oldroberta_data = pd.read_csv('roc_curve_oldRoberta.csv')
newroberta_data = pd.read_csv('roc_curve_newRoberta.csv')

fpr_distil = distil_data['FPR']
tpr_distil = distil_data['TPR']
fpr_oldroberta = oldroberta_data['FPR']
tpr_oldroberta = oldroberta_data['TPR']
fpr_newroberta = newroberta_data['FPR']
tpr_newroberta = newroberta_data['TPR']

plt.figure(figsize=(8, 6))
plt.plot(fpr_distil, tpr_distil, label='DistilBERT', color='blue')
plt.plot(fpr_oldroberta, tpr_oldroberta, label='RoBERTa-1', color='red')
plt.plot(fpr_newroberta, tpr_newroberta, label='RoBERTa-2', color='orange')
# plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Classifier')

plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

distil_data = pd.read_csv('pr_curve_distil.csv')
oldroberta_data = pd.read_csv('pr_curve_oldRoberta.csv')
newroberta_data = pd.read_csv('pr_curve_newRoberta.csv')

precision_distil = distil_data['Precision']
recall_distil = distil_data['Recall']
precision_oldroberta = oldroberta_data['Precision']
recall_oldroberta = oldroberta_data['Recall']
precision_newroberta = newroberta_data['Precision']
recall_newroberta = newroberta_data['Recall']

plt.figure(figsize=(8, 6))
plt.plot(recall_distil, precision_distil, label='DistilBERT', color='blue')
plt.plot(recall_oldroberta, precision_oldroberta, label='RoBERTa-1', color='red')
plt.plot(recall_newroberta, precision_newroberta, label='RoBERTa-2', color='orange')

plt.title('Precision-Recall Curve Comparison')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='lower left')
plt.grid(True)

plt.show()
```
